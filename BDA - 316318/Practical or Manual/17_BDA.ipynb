
# üê∑ Pig Script for Data Transformation

Apache **Pig** is used to process large datasets using **Pig Latin**.

---

## üìÇ Dataset (`data.txt`)

```text
1,Alice,HR,45000
2,Bob,IT,60000
3,Charlie,IT,70000
4,David,HR,50000
````

---

## üêñ Pig Script

### 1Ô∏è‚É£ Load Dataset into Pig

```pig
data = LOAD 'data.txt'
USING PigStorage(',')
AS (id:int, name:chararray, dept:chararray, salary:int);
```

---

### 2Ô∏è‚É£ Filter Records (Salary > 50,000)

```pig
filtered_data = FILTER data BY salary > 50000;
```

---

### 3Ô∏è‚É£ Group Data by Department

```pig
grouped_data = GROUP filtered_data BY dept;
```

---

### 4Ô∏è‚É£ Count Employees per Department

```pig
result = FOREACH grouped_data
GENERATE group AS department, COUNT(filtered_data) AS count;
```

---

### 5Ô∏è‚É£ Display Output

```pig
DUMP result;
```

---

## ‚úÖ Output

```text
(IT,2)
```
