# Running Basic Aggregation Queries on an Unstructured Dataset in MongoDB

---

## Step 1: Open MongoDB Shell

Open terminal or command prompt and run:

```bash
mongosh
```

---

## Step 2: Select Database

```bash
use reviewsDB
```

---

## Step 3: Create Collection

```bash
db.createCollection("product_reviews")
```

---

## Step 4: Insert Unstructured Dataset

```bash
db.product_reviews.insertMany([
  {
    userName: "Alice",
    rating: 4.5,
    review: "Good product",
    verifiedPurchase: true
  },
  {
    userName: "Bob",
    rating: 3.8,
    review: "Average quality",
    verifiedPurchase: false
  },
  {
    userName: "Charlie",
    rating: 4.9,
    review: "Excellent product",
    verifiedPurchase: true
  }
])
```

---

## Step 5: Verify Inserted Data

```bash
db.product_reviews.find().pretty()
```

---

## Step 6: Aggregation Query – Count Total Records

```bash
db.product_reviews.aggregate([
  { $count: "totalReviews" }
])
```

---

## Step 7: Aggregation Query – Calculate Average Rating

```bash
db.product_reviews.aggregate([
  {
    $group: {
      _id: null,
      averageRating: { $avg: "$rating" }
    }
  }
])
```
