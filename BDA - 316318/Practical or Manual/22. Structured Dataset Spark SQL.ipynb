## PySpark

`````python
# ==========================================
# Spark SQL - Random Structured Dataset
# Create Temporary View
# ==========================================

import numpy as np
import pandas as pd
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# -------------------------------
# 1ï¸âƒ£ Create Spark Session
# -------------------------------
spark = SparkSession.builder \
    .appName("SparkSQLTemporaryViewExample") \
    .getOrCreate()

# -------------------------------
# 2ï¸âƒ£ Generate Random Structured Data
# -------------------------------

np.random.seed(42)

n_rows = 100

data = {
    "employee_id": np.arange(1, n_rows + 1),
    "department": np.random.choice(["HR", "IT", "Sales", "Finance"], n_rows),
    "age": np.random.randint(22, 60, n_rows),
    "salary": np.random.randint(30000, 120000, n_rows),
    "experience_years": np.random.randint(1, 20, n_rows)
}

# Convert to Pandas DataFrame
pdf = pd.DataFrame(data)

# -------------------------------
# 3ï¸âƒ£ Convert Pandas â†’ Spark DataFrame
# -------------------------------
df = spark.createDataFrame(pdf)

print("=== Sample Data ===")
df.show(5)

# -------------------------------
# 4ï¸âƒ£ Create Temporary View
# -------------------------------
df.createOrReplaceTempView("employees")

print("\nTemporary View 'employees' created successfully!")

# -------------------------------
# 5ï¸âƒ£ Run Spark SQL Queries
# -------------------------------

# Query 1: Average salary by department
print("\n=== Average Salary by Department ===")
spark.sql("""
    SELECT department,
           ROUND(AVG(salary), 2) AS avg_salary
    FROM employees
    GROUP BY department
""").show()

# Query 2: Employees with salary > 80000
print("\n=== High Salary Employees (> 80000) ===")
spark.sql("""
    SELECT employee_id, department, salary
    FROM employees
    WHERE salary > 80000
""").show()

# Query 3: Count employees by department
print("\n=== Employee Count by Department ===")
spark.sql("""
    SELECT department,
           COUNT(*) AS total_employees
    FROM employees
    GROUP BY department
""").show()

# Query 4: Experienced employees (> 10 years)
print("\n=== Experienced Employees (> 10 years) ===")
spark.sql("""
    SELECT employee_id, department, experience_years
    FROM employees
    WHERE experience_years > 10
""").show()

# -------------------------------
# Stop Spark
# -------------------------------
spark.stop()
`````
