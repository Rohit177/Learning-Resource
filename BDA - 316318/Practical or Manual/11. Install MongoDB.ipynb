
# Install NoSQL Database (MongoDB) and Create Collections

---

## Software Required
- MongoDB Community Server
- MongoDB Shell (mongosh)
- Operating System: Windows / Linux / macOS

---

## MongoDB
MongoDB is a NoSQL, document-oriented database that stores data in JSON-like documents instead of tables and rows. It provides high performance, scalability, and flexibility.

---

## Installation Steps

### 1. Install MongoDB on Windows
1. Visit: https://www.mongodb.com/try/download/community
2. Download MongoDB Community Server (MSI).
3. Run the installer.
4. Select **Complete Setup**.
5. Enable **Install MongoDB as a Service**.
6. Finish installation.

---

## 2. Verify Installation

```bash
mongo
```

or

```bash
mongosh
```

---

## 3. Create Database

```bash
use schoolDB
```

---

## 4. Create Collection

### Automatic Collection Creation

```bash
db.students.insertOne({
  name: "John",
  age: 20,
  course: "Computer Science"
})
```

---

### Manual Collection Creation

```bash
db.createCollection("teachers")
```

---

## 5. Insert Multiple Documents

```bash
db.students.insertMany([
  { name: "Alice", age: 21, course: "IT" },
  { name: "Bob", age: 22, course: "AI" }
])
```

---

## 6. View Collections

```bash
show collections
```

---

## 7. View Documents

```bash
db.students.find().pretty()
```

---

## Result

MongoDB was successfully installed, and databases and collections were created.
