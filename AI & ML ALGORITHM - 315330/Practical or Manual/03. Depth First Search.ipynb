> 📖 **Description:**
**Depth First Search (DFS)** is an **uninformed search algorithm** used for traversing or searching graph and tree data structures. <br><br>
🔍 DFS explores a path as deeply as possible before backtracking to explore other paths. <br><br> 📊 It uses a **Stack (LIFO - Last In First Out)** data structure, either explicitly or through recursion.
---


```python

graph = {
  '1' : [],
  '2' : [],
  '4' : ['5','6'],
  '5' : ['7'],
  '6' : ['1','2'],
  '7' : []
}

# Sample view as node of graph
#        4
#    5       6
# 7       1     2


visited = []   # List for visited nodes.
stack = []     # Initialize a stack

def dfs(visited, stack, node): #function for DFS
  visited.append(node)
  stack.append(node)

  while stack:          # Creating loop to visit each node
    m = stack.pop()
    print (m, end = " ")

    for neighbour in graph[m]:
      if neighbour not in visited:
        visited.append(neighbour)
        stack.append(neighbour)

# Function call
print("\nFollowing is the Depth-First Search")
dfs(visited, stack, '4')
```
