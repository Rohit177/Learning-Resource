> 📖 **Description:**
**Bayes’ Theorem** is a fundamental concept in **Probability Theory** and **Machine Learning** that describes the probability of an event based on prior knowledge of conditions related to that event. <br><br>
🔍 It is used to update the probability of a hypothesis when new evidence becomes available. <br><br>
📊 In Bayes’ Theorem:
> - **P(A|B)** = Probability of event A occurring given that B has occurred.
> - **P(B|A)** = Probability of event B occurring given that A has occurred.
> - **P(A)** = Prior probability of event A.
> - **P(B)** = Probability of event B.
---

**Example:**
<br>
> **Disease test:** 1 in 1000 people have disease.

> 99% Accurate (true positive rate = 0.99).

> False positive rate is 5%.

```python
def bayes_theorem(pa, pba, pb):
    """
    Bayes' Theorem:
    P(A|B) = (P(B|A) * P(A)) / P(B)
    """
    pab = (pba * pa) / pb
    return pab


# Given:
pa = 0.01         # P(A)
pba = 0.95  # P(B|A)
pp = (pba * pa) + (0.05 * (1 - pa))  # Total positive

# Calculate
result = bayes_theorem(pa, pba, pp)

print(f"Positive test of disease: {result:.3f}")
```
