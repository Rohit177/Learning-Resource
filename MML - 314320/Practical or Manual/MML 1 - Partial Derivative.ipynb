> **Description:** A partial derivative measures how a function changes with respect to one variable while keeping all other variables constant. It is used for functions of multiple variables to see the effect of each variable individually. Essentially, it tells you the slope of the function along one direction in a multi-dimensional space.

```python
def f(x, y):
    return x**2 * y + 3*x*y**2

# First-order partial derivatives
def df_dx(x, y, h=1e-5):
    return (f(x + h, y) - f(x, y)) / h

def df_dy(x, y, h=1e-5):
    return (f(x, y + h) - f(x, y)) / h

# Second-order partial derivatives
def d2f_dx2(x, y, h=1e-5):
    return (f(x + h, y) - 2*f(x, y) + f(x - h, y)) / h**2

def d2f_dy2(x, y, h=1e-5):
    return (f(x, y + h) - 2*f(x, y) + f(x, y - h)) / h**2

# Mixed partial derivative
def d2f_dxdy(x, y, h=1e-5):
    return (f(x + h, y + h) - f(x + h, y) - f(x, y + h) + f(x, y)) / h**2

# Point to evaluate derivatives
x0, y0 = 1, 2

# Print each derivative on a separate line
print(f"First-order partial derivative = {df_dx(x0, y0)}\n")
print(f"First-order partial derivative of f with respect to y = {df_dy(x0, y0)}\n")
print(f"The second-order partial derivative of f with respect to x = {d2f_dx2(x0, y0)}\n")
print(f"The second-order partial derivative of f with respect to y = {d2f_dy2(x0, y0)}\n")
print(f"The mixed partial derivative of f with respect to x and y = {d2f_dxdy(x0, y0)}\n")
```
