> **Description:**  Vector algebra involves operations performed on vectors such as addition, subtraction, and scalar multiplication.
Vector addition and subtraction are done by adding or subtracting corresponding elements of vectors. Scalar multiplication multiplies each element of a vector by a constant value.

> **Resource:**
  
---
<center> </center>


```python
# Vector Addition
def vector_add(v1, v2):
    return [v1[i] + v2[i] for i in range(len(v1))]

# Vector Subtraction
def vector_sub(v1, v2):
    return [v1[i] - v2[i] for i in range(len(v1))]

# Scalar Multiplication
def scalar_mul(k, v):
    return [k * x for x in v]


# -------- Example Usage --------
v1 = [3, 4, 5]
v2 = [1, 2, 3]
k = 2  # scalar

print("Vector 1:", v1)
print("Vector 2:", v2)

print("\nVector Addition (v1 + v2):", vector_add(v1, v2))
print("Vector Subtraction (v1 - v2):", vector_sub(v1, v2))
print("Scalar Multiplication (k * v1):", scalar_mul(k, v1))
```
