> **Description:** Numerical differentiation is used to estimate the derivative of a function from given data values. It uses formulas like forward difference, backward difference, or central difference.

> **Resource:**

<center> </center>


```python
# Example data
x = [0, 1, 2, 3, 4]       # independent variable
y = [0, 1, 4, 9, 16]      # dependent variable f(x) = x^2

n = len(x)
h = x[1] - x[0]  # assuming uniform spacing

# Forward difference
def forward_diff(y, h):
    return [(y[i+1] - y[i])/h for i in range(n-1)] + [None]

# Backward difference
def backward_diff(y, h):
    return [None] + [(y[i] - y[i-1])/h for i in range(1, n)]

# Central difference
def central_diff(y, h):
    return [None] + [(y[i+1] - y[i-1])/(2*h) for i in range(1, n-1)] + [None]

# Compute derivatives
fwd = forward_diff(y, h)
bwd = backward_diff(y, h)
cen = central_diff(y, h)

# Display results
print("x\tf(x)\tForward\tBackward\tCentral")
for i in range(n):
    print(x[i], "\t", y[i], "\t", fwd[i], "\t", bwd[i], "\t", cen[i])
```
