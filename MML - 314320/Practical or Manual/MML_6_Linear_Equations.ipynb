```python
# Solve system of linear equations using Gauss Elimination

def solve_linear_system(A, b):
    n = len(A)

    # Form the augmented matrix [A | b]
    for i in range(n):
        A[i] = A[i] + [b[i]]

    # Forward elimination
    for i in range(n):
        # Pivot should not be zero; swap if needed
        if A[i][i] == 0:
            for r in range(i+1, n):
                if A[r][i] != 0:
                    A[i], A[r] = A[r], A[i]
                    break

        pivot = A[i][i]

        # Normalize pivot row
        for j in range(i, n+1):
            A[i][j] /= pivot

        # Eliminate below
        for r in range(i+1, n):
            factor = A[r][i]
            for c in range(i, n+1):
                A[r][c] -= factor * A[i][c]

    # Back substitution
    x = [0] * n
    for i in range(n-1, -1, -1):
        x[i] = A[i][n] - sum(A[i][j] * x[j] for j in range(i+1, n))

    return x


# ------------------ Example ------------------
A = [
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1,  2]
]

b = [8, -11, -3]

print("Solution of the system:")
solution = solve_linear_system([row[:] for row in A], b)
print(solution)
```

```python


# Example system:
# 2x + y - z = 8
# -3x - y + 2z = -11
# -2x + y + 2z = -3

import numpy as np

A = np.array([
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1, 2]
], dtype=float)

b = np.array([8, -11, -3], dtype=float)

solution = np.linalg.solve(A, b)

print("Solution of the system:")
for i, val in enumerate(solution, start=1):
    print(f"x{i} = {round(val, 6)}")
```
