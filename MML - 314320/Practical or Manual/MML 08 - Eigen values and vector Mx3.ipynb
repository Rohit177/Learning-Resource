```python
import numpy as np

# Function to compute eigenvalues and eigenvectors of 3x3 matrix
def eigen_3x3(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues, eigenvectors

# ----------- Example Matrix ------------
A = np.array([
    [2, -1, 0],
    [-1, 2, -1],
    [0, -1, 2]
])

values, vectors = eigen_3x3(A)

print("Matrix A:")
print(A)

print("\nEigenvalues:")
print(values)

print("\nEigenvectors (each column is an eigenvector):")
print(vectors)
```
