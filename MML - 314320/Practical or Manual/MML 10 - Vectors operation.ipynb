> **Description:**  Vector operations include dot product, cross product, and scalar triple product. The dot product of two vectors gives a scalar value and measures their similarity. The cross product produces a vector perpendicular to the given vectors. The scalar triple product gives the volume of the parallelepiped formed by three vectors.

```python
# Dot Product
def dot_product(v1, v2):
    return sum(v1[i] * v2[i] for i in range(len(v1)))

# Cross Product (for 3D vectors)
def cross_product(v1, v2):
    return [
        v1[1]*v2[2] - v1[2]*v2[1],
        v1[2]*v2[0] - v1[0]*v2[2],
        v1[0]*v2[1] - v1[1]*v2[0]
    ]

# Scalar Triple Product: a · (b × c)
def scalar_triple_product(a, b, c):
    return dot_product(a, cross_product(b, c))


# ----------- Example Usage -----------
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]

print("Vector a:", a)
print("Vector b:", b)
print("Vector c:", c)

print("\nDot product (a · b):", dot_product(a, b))
print("Cross product (a × b):", cross_product(a, b))
print("Scalar triple product (a · (b × c)):", scalar_triple_product(a, b, c))
```
