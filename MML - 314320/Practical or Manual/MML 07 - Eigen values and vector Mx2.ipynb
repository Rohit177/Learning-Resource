Eigenvectors are non-zero vectors that change only by a scalar factor when the matrix is applied to them.
For a 2Ã—2 matrix, a program computes eigenvalues from the characteristic equation and then finds the corresponding eigenvectors.

```python
import numpy as np

# Function to compute eigenvalues and eigenvectors
def eigen_2x2(matrix):
    values, vectors = np.linalg.eig(matrix)
    return values, vectors

# ---------- Example ----------
A = np.array([[4, 2],
              [1, 3]])

eigen_values, eigen_vectors = eigen_2x2(A)

print("Matrix A:")
print(A)

print("\nEigenvalues:")
print(eigen_values)

print("\nEigenvectors (column-wise):")
print(eigen_vectors)
```
