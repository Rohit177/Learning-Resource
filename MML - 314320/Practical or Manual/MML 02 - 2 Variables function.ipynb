> **Description:**  The maximum and minimum of a function for two variables refer to the highest and lowest values of a function involving two independent variables. These values occur at specific points where the function changes its behavior. They are commonly found by analyzing critical points and boundaries. This concept is widely used in optimization problems.

```python
import numpy as np
from scipy.optimize import minimize

def function_to_optimize(variables):
    x, y = variables
    return x**2 + y**2 - 4*x + 3*y

# Find minimum of the function
initial_guess = [0, 0]

min_result = minimize(function_to_optimize, initial_guess)


# Find maximum of the function by minimizing the negative

def negative_function(variables):
    return -function_to_optimize(variables)

max_result = minimize(negative_function, initial_guess)


# Display Results

print("Minimum value of f(x, y):", min_result.fun)
print("Occurs at (x, y):", min_result.x)

print("\nMaximum value of f(x, y):", -max_result.fun)
print("Occurs at (x, y):", max_result.x)
```
