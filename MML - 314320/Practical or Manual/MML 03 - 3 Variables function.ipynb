> **Description:**  The maximum and minimum of a function for three variables are the highest and lowest values the function can attain. These occur at critical points where all partial derivatives are zero or at boundary points. By analyzing these points, the nature of the function can be determined. This method is commonly used in optimization problems.

```python
import numpy as np
from scipy.optimize import minimize

# ----------------------------------------------------------
# Function of three variables: f(x, y, z)
# Example: f(x, y, z) = x^2 + y^2 + z^2 - 4x + 3y + 2z
# ----------------------------------------------------------
def function_to_optimize(variables):
    x, y, z = variables
    return x**2 + y**2 + z**2 - 4*x + 3*y + 2*z


# ----------------------------------------------------------
# Initial guess for optimization
# ----------------------------------------------------------
initial_guess = [0, 0, 0]

# ----------------------------------------------------------
# Find minimum of the function
# ----------------------------------------------------------
min_result = minimize(function_to_optimize, initial_guess)


# ----------------------------------------------------------
# Find maximum by minimizing the negative function
# ----------------------------------------------------------
def negative_function(variables):
    return -function_to_optimize(variables)

max_result = minimize(negative_function, initial_guess)


# ----------------------------------------------------------
# Display results
# ----------------------------------------------------------
print("========== RESULTS ==========")

print("Minimum value of f(x, y, z):", min_result.fun)
print("Occurs at (x, y, z):", min_result.x)

print("\nMaximum value of f(x, y, z):", -max_result.fun)
print("Occurs at (x, y, z):", max_result.x)
```
