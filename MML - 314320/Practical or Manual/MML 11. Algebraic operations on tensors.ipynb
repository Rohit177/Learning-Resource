> **Description:** Tensor algebra involves mathematical operations on tensors of the same order and dimensions. Addition and subtraction of tensors are performed by adding or subtracting corresponding elements.

```python
# Recursive function to add two tensors
def tensor_add(A, B):
    if isinstance(A, list) and isinstance(B, list):
        return [tensor_add(a, b) for a, b in zip(A, B)]
    else:
        return A + B

# Recursive function to subtract two tensors
def tensor_sub(A, B):
    if isinstance(A, list) and isinstance(B, list):
        return [tensor_sub(a, b) for a, b in zip(A, B)]
    else:
        return A - B

# Example tensors
tensor1 = [
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
]

tensor2 = [
    [[8, 7], [6, 5]],
    [[4, 3], [2, 1]]
]

print("Tensor 1:")
print(tensor1)

print("\nTensor 2:")
print(tensor2)

# Tensor Addition
sum_tensor = tensor_add(tensor1, tensor2)
print("\nTensor Addition (Tensor1 + Tensor2):")
print(sum_tensor)

# Tensor Subtraction
diff_tensor = tensor_sub(tensor1, tensor2)
print("\nTensor Subtraction (Tensor1 - Tensor2):")
print(diff_tensor)
```
