> **Description:** Numerical integration approximates the area under a curve using given data points. The Trapezoidal Rule divides the area into trapezoids and sums their areas. It is simple, works with discrete data, and gives good accuracy for smooth functions or small intervals.

```python
# Example data
x = [0, 1, 2, 3, 4]       # independent variable
y = [0, 1, 4, 9, 16]      # dependent variable f(x) = x^2

n = len(x)
h = x[1] - x[0]  # assuming uniform spacing

# Trapezoidal Rule
def trapezoidal_rule(x, y):
    n = len(x)
    integral = 0
    for i in range(n-1):
        integral += (y[i] + y[i+1]) * (x[i+1] - x[i]) / 2
    return integral

# Compute integral
result = trapezoidal_rule(x, y)

print("Trapezoidal Rule Approximation of Integral:", result)
```
